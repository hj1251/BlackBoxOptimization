```python
import ast
import os

BASE_DIR = os.path.join(os.getcwd(), "initial_data")

def load_week_data(week, base_dir="initial_data"):
    with open(f"{base_dir}/inputs_week{week}.txt", "r") as f:
        X_new_list = eval(f.read(), {"array": np.array})

    with open(f"{base_dir}/outputs_week{week}.txt", "r") as f:
        y_new_list = eval(f.read())

    return X_new_list, y_new_list

```

```python
def append_new_point(X, y, x_new, y_new):
    X_updated = np.vstack([X, np.array(x_new).reshape(1, -1)])
    y_updated = np.append(y, y_new)
    return X_updated, y_updated
```

At first glance, the predictive standard deviation (std) at the observed training points is extremely small, 
which may give the impression that the Gaussian Process model is overconfident in its predictions.

To further investigate this, I evaluated the model uncertainty over a large set of randomly sampled points 
in the input space. The maximum predicted standard deviation over these unseen points is significantly larger 
than the std at the training locations.

This indicates that the model is only confident near the observed data, while substantial uncertainty remains 
in unexplored regions of the input space. Therefore, the small std at training points is expected and does not 
imply that the model has fully converged.

As a result, the optimization process should continue to encourage exploitation of uncertain regions rather 
than relying solely on exploitation of the current predictions.

At this stage, I choose to use the Upper Confidence Bound (UCB) acquisition function rather than Expected Improvement (EI).

From the current observations, the overall scale of the output values is very small, and even the best observed y so far is close to zero. This makes EI less suitable, as EI explicitly measures the expected improvement relative to the current maximum y. When the current best y is very small, EI tends to become overly conservative and biased towards exploitation around already-sampled regions.

In contrast, UCB directly balances the predicted mean and the model uncertainty. By assigning additional weight to the predictive standard deviation, UCB encourages exploration in regions where the model is still uncertain, even if the predicted mean is not currently high.

Since the current model is fitted on a very limited number of data points and the global structure of the function is still unclear, prioritising exploration is more appropriate at this stage. Therefore, UCB is selected to guide the next query point selection.


```python
X_cand = np.random.rand(5000, X1.shape[1])
mu, std = gpr1.predict(X_cand, return_std=True)

kappa = 2.5
ucb = mu + kappa * std

# Select the next query point
x_next = X_cand[np.argmax(ucb)]

print(x_next)
```

Although the Gaussian Process model for Function 2 fits the observed data well and shows low uncertainty around the sampled points, the predictive uncertainty remains relatively high in unexplored regions of the input space.

Therefore, I choose the Upper Confidence Bound (UCB) acquisition function to encourage exploration. By explicitly incorporating the predictive standard deviation, UCB allows the optimisation process to explore regions where the model is less certain, rather than focusing solely on exploiting the current best observed values.

```python
def ucb_acquisition(X, model, beta=2.0):
    mean, std = model.predict(X, return_std=True)
    return mean + beta * std
```

For Function 5, the observed y distribution is highly skewed with a clear outlier. Instead of performing uniform random exploration over the entire input space,I focus on a local region around the current best observed point. By probing the GP's predictive mean and uncertainty in this neighbourhood, I can assess whether the model remains uncertain near the promising region.

A local UCB strategy is then applied to balance exploitation of the high-y region and exploration driven by remaining uncertainty.


```python
best_idx = np.argmax(y5)
x5_best = X5[best_idx]
y5_best = y5[best_idx]

print("Best observed y:", y5_best)
print("x at best y:", x5_best)
```

```python
def sample_local_region(x_center, n_samples=5000, radius=0.05):
    d = len(x_center)
    noise = np.random.normal(0, radius, size=(n_samples, d))
    X_local = x_center + noise
    return np.clip(X_local, 0.0, 1.0)

X5_local = sample_local_region(
    x_center=x5_best,
    n_samples=5000,
    radius=0.05
)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, WhiteKernel, ConstantKernel

# ---- Function 5: define + fit gpr5 ----
d5 = X5.shape[1]

kernel5 = ConstantKernel(1.0, (1e-3, 1e3)) * Matern(
    length_scale=np.ones(d5),
    length_scale_bounds=(1e-2, 1e2),
    nu=2.5
) + WhiteKernel(
    noise_level=1e-5,
    noise_level_bounds=(1e-8, 1e-1)
)

gpr5 = GaussianProcessRegressor(
    kernel=kernel5,
    normalize_y=True,
    alpha=1e-8,
    n_restarts_optimizer=8,
    random_state=0
)

gpr5.fit(X5, y5)
print("gpr5 fitted. kernel:", gpr5.kernel_)
```

```python
mu_local, std_local = gpr5.predict(X5_local, return_std=True)

print("Local mean range:", mu_local.min(), mu_local.max())
print("Local std range:", std_local.min(), std_local.max())
print("Max local std:", std_local.max())
```

```python
rng = np.random.default_rng(42)
n_candidates = 30000
X_candidates_5 = rng.random((n_candidates, X5.shape[1]))

mu5, std5 = gpr5.predict(X_candidates_5, return_std=True)

kappa = 2.5  # early stage -> more exploration
ucb5 = mu5 + kappa * std5

best_idx_5 = np.argmax(ucb5)
x5_next = X_candidates_5[best_idx_5]

print("mu/std at x5_next:", mu5[best_idx_5], std5[best_idx_5])
```

Although a global candidate set (uniform sampling over the entire input space) provides a baseline estimate of model uncertainty, the comparison shows that the predictive standard deviation within a local region around high-performing points is substantially larger. This indicates that, despite having a promising predicted mean, the local neighbourhood remains under-explored and the model is still highly uncertain there. Therefore, instead of selecting the next point from a purely random global candidate set, we restrict the search to this local region and apply UCB-based selection to balance exploitation (high predicted mean) and exploration (high local uncertainty).

As a result, the next query point is chosen from the local candidate set rather than from random global sampling.


```python
def propose_next_point_local_ucb(
    gpr,
    X, y,
    n_candidates=30000,
    radius=0.08,
    kappa=2.5,
    random_state=42,
):
  
    rng = np.random.default_rng(random_state)

    best_idx = np.argmax(y)
    x_best = X[best_idx]
    y_best = y[best_idx]

    # local box bounds, clipped to [0,1]
    lower = np.clip(x_best - radius, 0.0, 1.0)
    upper = np.clip(x_best + radius, 0.0, 1.0)

    # sample candidates uniformly inside the local box
    X_local = lower + (upper - lower) * rng.random((n_candidates, X.shape[1]))

    # GP predictive mean/std on candidates
    mu, std = gpr.predict(X_local, return_std=True)

    # UCB acquisition
    ucb = mu + kappa * std
    idx_next = np.argmax(ucb)
    x_next = X_local[idx_next]

    info = {
        "x_best": x_best,
        "y_best": y_best,
        "mu_next": mu[idx_next],
        "std_next": std[idx_next],
        "ucb_next": ucb[idx_next],
        "mu_range": (mu.min(), mu.max()),
        "std_range": (std.min(), std.max()),
        "lower": lower,
        "upper": upper,
    }
    return x_next, info
```

```python
x5_next, info5 = propose_next_point_local_ucb(
    gpr=gpr5,
    X=X5,
    y=y5,
    n_candidates=30000,
    radius=0.08,   
    kappa=2.5,    
    random_state=42,
)

print("Current best y:", info5["y_best"])
print("Proposed x5_next:", x5_next)
print("Predicted mu/std at x5_next:", info5["mu_next"], info5["std_next"])
print("Local mu range:", info5["mu_range"])
print("Local std range:", info5["std_range"])
print("Local box lower/upper:", info5["lower"], info5["upper"])
```

By randomly selecting a single tree from the ensemble and optimising with respect to that sampled model, the method naturally balances exploration and exploitation without relying on explicit tuning of an exploration parameter.

```python
rng = np.random.default_rng(42)

# Global candidate set (high-dimensional -> sample more)
X_cand = rng.random((20000, X6.shape[1]))

# Thompson sampling: sample one tree
tree = rng.choice(rf6.estimators_)

# Predict using the sampled tree
pred = tree.predict(X_cand)

# Select the best point under this sampled model
x6_next = X_cand[np.argmax(pred)]

print("Proposed x6_next (Thompson Sampling):", x6_next)
print("Predicted y under sampled tree:", pred.max())
```

```python
import numpy as np

def get_next_candidate(
    gpr,
    X,
    n_cand=5000,
    kappa=1.5,
    random_state=None,
    min_dist=None
):
    rng = np.random.default_rng(random_state)
    X = np.asarray(X)
    dim = X.shape[1]

    X_cand = rng.random((n_cand, dim))

    if min_dist is not None and X.shape[0] > 0:
        d2 = ((X_cand[:, None, :] - X[None, :, :]) ** 2).sum(axis=2)
        min_d = np.sqrt(d2.min(axis=1))
        keep = min_d >= float(min_dist)
        if np.any(keep):
            X_cand = X_cand[keep]
        if X_cand.shape[0] == 0:
            X_cand = rng.random((n_cand, dim))

    mu, std = gpr.predict(X_cand, return_std=True)

    std = np.maximum(std, 1e-12)


    ucb = mu + kappa * std
    idx_next = int(np.argmax(ucb))

    x_next = X_cand[idx_next:idx_next + 1] 
    mu_next = float(mu[idx_next])
    std_next = float(std[idx_next])
    ucb_next = float(ucb[idx_next])

    return x_next, mu_next, std_next, ucb_next
```

```python
def propose_func3(
    gpr,
    X_obs,
    n_cand=8000,
    kappa=1.0,
    random_state=0,
    bias_strength=2.0
):
    rng = np.random.default_rng(random_state)
    dim = X_obs.shape[1]
    assert dim == 3

    X_cand = rng.random((n_cand, dim))

    a = float(bias_strength)
    u0 = rng.random(n_cand)
    u1 = rng.random(n_cand)
    u2 = rng.random(n_cand)

    X_cand[:, 0] = 1.0 - (u0 ** a)   # pushes toward 1
    X_cand[:, 1] = 1.0 - (u1 ** a)   # pushes toward 1
    X_cand[:, 2] = (u2 ** a)         # pushes toward 0

    # GP predictions
    mu, std = gpr.predict(X_cand, return_std=True)
    std = np.maximum(std, 1e-12)

    # UCB acquisition (more exploitation than early weeks)
    ucb = mu + kappa * std
    idx = int(np.argmax(ucb))

    x_next = X_cand[idx:idx+1]
    return x_next, float(mu[idx]), float(std[idx]), float(ucb[idx])

x_next3, mu3, std3, ucb3 = propose_func3(gpr3, X3, kappa=1.0, random_state=0, bias_strength=2.0)
print("x_next3:", x_next3.round(6))
print("mu/std/ucb:", mu3, std3, ucb3)
```

```python
def rf_predict_mean_std(rf_model, X):
    """
    Estimate predictive mean and uncertainty for RandomForestRegressor.
    - mean: average prediction across trees
    - std : standard deviation across trees (uncertainty proxy)
    """
    tree_preds = np.stack([est.predict(X) for est in rf_model.estimators_], axis=0)
    mu = tree_preds.mean(axis=0)
    std = tree_preds.std(axis=0, ddof=1)
    return mu, std

n_cand = 30000
dim = X6.shape[1]
rng = np.random.default_rng(42) 
X_cand = rng.random((n_cand, dim))

mu6, std6 = rf_predict_mean_std(rf6, X_cand)

print("Candidate mu range:", float(mu6.min()), float(mu6.max()))
print("Candidate std range:", float(std6.min()), float(std6.max()))

p = 0.10 
std_thr = np.quantile(std6, 1 - p)
mask = std6 >= std_thr

X_unc = X_cand[mask]
mu_unc = mu6[mask]
std_unc = std6[mask]

idx_local = int(np.argmax(mu_unc))
x6_next = X_unc[idx_local:idx_local+1]

print("Next candidate (Function 6, uncertainty-gated):", np.round(x6_next, 6))
print("Predicted std at x_next:", float(std_unc[idx_local]))
print("Predicted mean at x_next:", float(mu_unc[idx_local]))
print("Current best y6:", float(np.max(y6)))
```

```python
def pick_next_x3_coverage(X3, split=0.5, n=2000, seed=0, verbose=True):
    """
    Coverage strategy for Function 3:
    1) Split each dimension at `split` -> 2x2x2 regions
    2) Find least explored region(s)
    3) Sample candidates uniformly inside those region(s)
    4) Pick point that maximizes distance to nearest existing sample (maximin)
    """
    X3_arr = np.asarray(X3, dtype=float)

    i0 = (X3_arr[:, 0] > split).astype(int)
    i1 = (X3_arr[:, 1] > split).astype(int)
    i2 = (X3_arr[:, 2] > split).astype(int)

    grid = np.zeros((2, 2, 2), dtype=int)
    for a, b, c in zip(i0, i1, i2):
        grid[a, b, c] += 1

    min_count = grid.min()
    cells = np.argwhere(grid == min_count)

    rng = np.random.default_rng(seed)

    def sample_cell(cell):
        lo = np.array([
            split if cell[0] else 0.0,
            split if cell[1] else 0.0,
            split if cell[2] else 0.0
        ], dtype=float)
        hi = np.array([
            1.0 if cell[0] else split,
            1.0 if cell[1] else split,
            1.0 if cell[2] else split
        ], dtype=float)
        return lo + (hi - lo) * rng.random((n, 3))

    best_x = None
    best_dist = -1.0
    best_cell = None

    for cell in cells:
        P = sample_cell(cell)

        diff = P[:, None, :] - X3_arr[None, :, :]
        dist = np.sqrt((diff**2).sum(axis=2))
        min_dist = dist.min(axis=1)

        idx = np.argmax(min_dist)
        if min_dist[idx] > best_dist:
            best_dist = float(min_dist[idx])
            best_x = P[idx]
            best_cell = cell.copy()

    if verbose:
        print("=== 2x2x2 region counts ===")
        for a in range(2):
            for b in range(2):
                for c in range(2):
                    print(
                        f"x0 {'>0.5' if a else '<=0.5'}, "
                        f"x1 {'>0.5' if b else '<=0.5'}, "
                        f"x2 {'>0.5' if c else '<=0.5'} :",
                        grid[a, b, c]
                    )
        print("\nLeast explored regions:", cells.tolist(), "(min count =", int(min_count), ")")
        print("\n=== Suggested next point ===")
        print("region:", best_cell.tolist())
        print("x_next:", np.round(best_x, 6))
        print("distance to nearest sample:", best_dist)

    return best_x, best_cell, best_dist, grid
```

To verify whether further local refinement is worthwhile, we examine the GP predictive uncertainty around the current best observation.

Although the Gaussian Process typically has very low variance at observed training points (e.g. at `x_best`), the surrounding region may still be uncertain. Therefore, we sample a set of candidate points in a small neighbourhood around `x_best` and evaluate the predictive standard deviation.

By computing the distribution of `std` at increasing radii (e.g. `r ≤ 0.05`, `0.10`, `0.15`, `0.20`), we can distinguish between two situations:

- Well-learned local region: both the best point and its neighbourhood have low predictive uncertainty.
- Partially explored basin: the best point is certain, but nearby points still have high uncertainty, suggesting the local landscape is not yet fully learned.

This diagnostic helps determine whether the optimisation should continue local refinement (trust-region search) or shift back to broader exploration.


```python
X5_arr = np.asarray(X5, float)
y5_arr = np.asarray(y5, float).reshape(-1)

best_idx = int(np.argmax(y5_arr))
x_best = X5_arr[best_idx]

rng = np.random.default_rng(0)
radius = 0.08
X_near = np.clip(x_best + rng.normal(0.0, radius, size=(8000, X5_arr.shape[1])), 0.0, 1.0)
_, std_near = gpr5.predict(X_near, return_std=True)
std_near = std_near.reshape(-1)

dist = np.linalg.norm(X_near - x_best[None, :], axis=1)

for r in [0.05, 0.10, 0.15, 0.20]:
    m = dist <= r
    if m.sum() == 0:
        print(f"r<={r}: no points")
        continue
    q = np.quantile(std_near[m], [0.5, 0.9, 0.95, 0.99])
    print(f"r<={r}: n={m.sum()}, std median/p90/p95/p99 = {q}")
```

A diagnostic check of the Gaussian Process predictive uncertainty around the current best point shows that while the variance at the observed maximum itself is very small (as expected for training data), the surrounding neighbourhood still exhibits relatively high predictive standard deviation. This suggests that the local structure of the high-value basin has not yet been fully learned by the model.

Therefore, the optimisation continues with a trust-region refinement strategy, selecting the next point within the neighbourhood of the current best to further explore this promising region rather than prematurely switching to global exploration.


```python
X5_arr = np.asarray(X5, dtype=float)
y5_arr = np.asarray(y5, dtype=float).reshape(-1)
n, d = X5_arr.shape

best_idx = int(np.argmax(y5_arr))
x_best = X5_arr[best_idx].copy()

dist_to_best = np.linalg.norm(X5_arr - x_best[None, :], axis=1)
dist_sorted = np.sort(dist_to_best)
nn3 = float(dist_sorted[2]) if len(dist_sorted) > 2 else float(dist_sorted[1]) if len(dist_sorted) > 1 else 0.1
radius = float(np.clip(2.5 * nn3, 0.02, 0.10))

rng = np.random.default_rng(0)

min_sep = 0.06  

def sample_local_resample(center, sigma, n_samples, max_tries=50):
    X = rng.normal(loc=center, scale=sigma, size=(n_samples, d))

    mask = (X >= 0.0).all(axis=1) & (X <= 1.0).all(axis=1)
    out = X[mask]
    tries = 0
    while out.shape[0] < n_samples and tries < max_tries:
        need = n_samples - out.shape[0]
        X_new = rng.normal(loc=center, scale=sigma, size=(need, d))
        mask_new = (X_new >= 0.0).all(axis=1) & (X_new <= 1.0).all(axis=1)
        if mask_new.any():
            out = np.vstack([out, X_new[mask_new]])
        tries += 1
        
    if out.shape[0] < n_samples:
        need = n_samples - out.shape[0]
        X_fb = np.clip(rng.normal(loc=center, scale=sigma, size=(need, d)), 0.0, 1.0)
        out = np.vstack([out, X_fb])
    return out[:n_samples]

n_samples = 20000
X_local = sample_local_resample(x_best, radius, n_samples)

dist_mat = np.linalg.norm(X_local[:, None, :] - X5_arr[None, :, :], axis=2)
min_dist = dist_mat.min(axis=1)
X_local = X_local[min_dist >= min_sep]

if X_local.shape[0] < 2000:
    radius2 = min(0.15, radius * 1.8)
    X_local2 = sample_local_resample(x_best, radius2, n_samples)
    dist_mat2 = np.linalg.norm(X_local2[:, None, :] - X5_arr[None, :, :], axis=2)
    min_dist2 = dist_mat2.min(axis=1)
    X_local2 = X_local2[min_dist2 >= min_sep]
    if X_local2.shape[0] > X_local.shape[0]:
        X_local = X_local2
        radius = radius2

kappa = 0.3
mu, std = gpr5.predict(X_local, return_std=True)
std = np.maximum(std, 1e-12)
ucb = mu + kappa * std

idx = int(np.argmax(ucb))
x_next = X_local[idx]
x_next_submit = np.round(x_next, 6)

print("x_best:", np.round(x_best, 6))
print("radius used:", radius)
print("min_sep:", min_sep)
print("candidate pool:", X_local.shape[0])
print("x_next (submit):", x_next_submit)
print("distance to nearest existing:", float(np.min(np.linalg.norm(X5_arr - x_next[None, :], axis=1))))
print("pred mu/std/ucb:", float(mu[idx]), float(std[idx]), float(ucb[idx]))
```

```python
def find_largest_hole(X_obs, n_probe=60000, seed=0, batch=5000):

    rng = np.random.default_rng(seed)
    X_obs = np.asarray(X_obs, dtype=float)
    n, d = X_obs.shape

    best_x = None
    best_min_dist = -1.0
    all_min_d = []

    left = int(n_probe)
    while left > 0:
        b = min(batch, left)
        left -= b

        P = rng.random((b, d))

        d2 = ((P[:, None, :] - X_obs[None, :, :]) ** 2).sum(axis=2)
        min_d = np.sqrt(d2.min(axis=1))  # (b,)

        all_min_d.append(min_d)

        i = int(np.argmax(min_d))
        if float(min_d[i]) > best_min_dist:
            best_min_dist = float(min_d[i])
            best_x = P[i].copy()

    all_min_d = np.concatenate(all_min_d, axis=0)
    stats = {
        "min_dist_mean": float(all_min_d.mean()),
        "min_dist_p95": float(np.quantile(all_min_d, 0.95)),
        "min_dist_max": float(best_min_dist),
    }
    return best_x, best_min_dist, stats


def propose_next_by_largest_hole(X_obs, n_probe=60000, seed=0, batch=5000,
                                jitter=0.01, min_dist=0.0):

    rng = np.random.default_rng(seed)
    X_obs = np.asarray(X_obs, dtype=float)
    d = X_obs.shape[1]

    x_hole, hole_dist, stats = find_largest_hole(X_obs, n_probe=n_probe, seed=seed, batch=batch)

    x_next = np.clip(x_hole + rng.normal(0.0, jitter, size=d), 0.0, 1.0)

    if min_dist > 0:
        for _ in range(30):
            d_near = float(np.min(np.linalg.norm(X_obs - x_next[None, :], axis=1)))
            if d_near >= min_dist:
                break
            x_next = np.clip(x_hole + rng.normal(0.0, jitter, size=d), 0.0, 1.0)

    x_next_submit = np.round(x_next, 6)

    details = {
        "x_hole_raw": np.round(x_hole, 6),
        "hole_nn_dist": float(hole_dist),
        "stats": stats,
        "x_next_submit": x_next_submit,
        "submit_string": ", ".join([f"{v:.6f}" for v in x_next_submit]),
        "dist_to_nearest_existing": float(np.min(np.linalg.norm(X_obs - x_next[None, :], axis=1))),
    }
    return x_next_submit, details

```

```python
def check_grid_coverage(X, split=0.5):
    
    X = np.asarray(X)
    
    i0 = (X[:,0] > split).astype(int)
    i1 = (X[:,1] > split).astype(int)
    i2 = (X[:,2] > split).astype(int)

    grid = np.zeros((2,2,2), dtype=int)

    for a,b,c in zip(i0,i1,i2):
        grid[a,b,c] += 1

    print("=== Grid coverage (2x2x2) ===\n")

    for a in range(2):
        for b in range(2):
            for c in range(2):

                region = (
                    f"x0 {'>0.5' if a else '<=0.5'}, "
                    f"x1 {'>0.5' if b else '<=0.5'}, "
                    f"x2 {'>0.5' if c else '<=0.5'}"
                )

                count = grid[a,b,c]

                if count == 0:
                    status = " not covered"
                else:
                    status = " covered"

                print(f"{region} : {count} samples   {status}")

    return grid
```

```python
import numpy as np
from itertools import product

def check_grid_coverage(X, bins=2):
    X = np.asarray(X)
    n, d = X.shape
    
    grid = {}
    
    idx = np.floor(X * bins).astype(int)
    idx[idx == bins] = bins - 1
    
    for row in idx:
        key = tuple(row)
        grid[key] = grid.get(key, 0) + 1
    
    print(f"\n=== {d}D Grid coverage ({bins}^{d} regions) ===")
    
    for region in product(range(bins), repeat=d):
        count = grid.get(region, 0)
        status = "covered" if count > 0 else "EMPTY"
        print(region, ":", count, status)
        
    return grid
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

X = np.asarray(X8, dtype=float)
y = np.asarray(y8, dtype=float).reshape(-1)

n, d = X.shape
print("n =", n, "dim =", d)

kernel = (
    C(1.0, (1e-3, 1e3)) *
    RBF(length_scale=np.ones(d), length_scale_bounds=(1e-2, 1e2)) +
    WhiteKernel(noise_level=1e-5, noise_level_bounds=(1e-10, 1e0))
)

gpr = GaussianProcessRegressor(
    kernel=kernel,
    normalize_y=True,
    n_restarts_optimizer=5,
    random_state=0
)

gpr.fit(X, y)
y_pred_gp = gpr.predict(X)

rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf.fit(X, y)
y_pred_rf = rf.predict(X)

rmse_gp = np.sqrt(mean_squared_error(y, y_pred_gp))
rmse_rf = np.sqrt(mean_squared_error(y, y_pred_rf))

print("\n=== Training RMSE ===")
print("GP RMSE:", rmse_gp)
print("RF RMSE:", rmse_rf)

print("\n=== LOOCV (leave-one-out) ===")

gp_errors = []
rf_errors = []

for i in range(n):
    X_train = np.delete(X, i, axis=0)
    y_train = np.delete(y, i)

    X_test = X[i].reshape(1, -1)
    y_true = y[i]

    # GP
    gpr_i = GaussianProcessRegressor(
        kernel=kernel,
        normalize_y=True,
        n_restarts_optimizer=2,
        random_state=0
    )
    gpr_i.fit(X_train, y_train)
    y_gp = gpr_i.predict(X_test)[0]

    # RF
    rf_i = RandomForestRegressor(
        n_estimators=200,
        random_state=42,
        n_jobs=-1
    )
    rf_i.fit(X_train, y_train)
    y_rf = rf_i.predict(X_test)[0]

    gp_errors.append((y_gp - y_true) ** 2)
    rf_errors.append((y_rf - y_true) ** 2)

rmse_gp_cv = np.sqrt(np.mean(gp_errors))
rmse_rf_cv = np.sqrt(np.mean(rf_errors))

print("GP LOOCV RMSE:", rmse_gp_cv)
print("RF LOOCV RMSE:", rmse_rf_cv)

print("\n=== Decision ===")
if rmse_gp_cv < rmse_rf_cv:
    print("👉 GP")
else:
    print("👉 RF")
```

In this stage of the optimisation, a Gaussian Process (GP) surrogate model is preferred over a Random Forest (RF) due to its superior ability to model smooth underlying structures and provide well-calibrated uncertainty estimates. Although both models can fit the observed data, the GP demonstrates stronger generalisation capability, as evidenced by its lower training error and its ability to identify improved candidate points in previous iterations. In contrast, the RF model, while robust, lacks a principled uncertainty quantification, which limits its effectiveness in guiding exploration in high-dimensional continuous spaces.

Furthermore, analysis of the observed high-value points indicates that they are widely scattered across the input space rather than forming a tight local cluster. This suggests that the optimisation has not yet converged to a single promising basin of attraction. As a result, a purely local search strategy is unlikely to yield further improvement. Instead, an exploration-oriented approach is required to better understand the global structure of the objective function.

Therefore, a GP-based Upper Confidence Bound (UCB) acquisition function with an increased exploration parameter is adopted. By increasing the exploration weight, the algorithm prioritises regions with higher predictive uncertainty, allowing it to discover potentially better regions that have not yet been sufficiently sampled. This strategy balances exploitation of known high-performing areas with broader exploration, which is particularly important given the limited number of remaining evaluations.

```python
loo = LeaveOneOut()

gp_errors = []
rf_errors = []

for train_idx, test_idx in loo.split(X4):
    X_train, X_test = X4[train_idx], X4[test_idx]
    y_train, y_test = y4[train_idx], y4[test_idx]

    # ===== GP =====
    gp = GaussianProcessRegressor(
        kernel=C(1.0) * Matern(nu=2.5) + WhiteKernel(),
        normalize_y=True,
        random_state=0
    )
    gp.fit(X_train, y_train)
    y_pred_gp = gp.predict(X_test)
    gp_errors.append((y_pred_gp - y_test)**2)

    rf = RandomForestRegressor(n_estimators=100, random_state=0)
    rf.fit(X_train, y_train)
    y_pred_rf = rf.predict(X_test)
    rf_errors.append((y_pred_rf - y_test)**2)

gp_mse = np.mean(gp_errors)
rf_mse = np.mean(rf_errors)

print("GP LOOCV MSE:", gp_mse)
print("RF LOOCV MSE:", rf_mse)
```
